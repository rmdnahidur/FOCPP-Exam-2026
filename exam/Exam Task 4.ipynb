```python
#Task 4: 

def check_range(lst):
    # Iterate through each number in the list
    for num in lst:
        # If any number is less than 0 or greater than 100, return False
        if num < 0 or num > 100:
            return False
    # If all numbers are in range, return True
    return True
test_list1 = [10, 50, 100]  
test_list2 = [5, 150, 99]   
print(f"List {test_list1}: {check_range(test_list1)}")  # Should print True
print(f"List {test_list2}: {check_range(test_list2)}")  # Should print False
```
