```python
#Task 5:

unique_strings = set()

while True:
    # Prompt the user for a string
    user_input = input("Enter a string (or press Enter to finish): ")
    # If the input is empty, break out of the loop
    if user_input == "":
        break
    # Add the input to the set (duplicates are ignored)
    unique_strings.add(user_input)

# Display the count of unique strings
print(f"Number of unique strings entered: {len(unique_strings)}")
```
